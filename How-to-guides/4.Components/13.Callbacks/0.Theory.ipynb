# Callbacks
LangChain provides a callbacks system that allows you to hook into the various stages of your LLM application. 

This is useful for logging, monitoring, streaming, and other tasks.

You can subscribe to these events by using the callbacks argument available throughout the API. 

This argument is list of handler objects, which are expected to implement one or more of the methods described below in more detail.




https://python.langchain.com/v0.2/docs/concepts/#callbacks
